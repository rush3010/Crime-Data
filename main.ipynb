```python
df3.isnull().sum()
```

```python
#Quantile based flooring and capping
#the outlier is capped at a certain value above the 90th percentile value or floored at a factor below the 10th percentile value.
# Computing 10th, 90th percentiles and replacing the outliers

#FOR Mileage column
def floor_cap(x):
    tenth_percentile = np.percentile(x, 10)
    ninetieth_percentile = np.percentile(x, 90)
    print("10th percentile is ",tenth_percentile)
    print("90th percentile is ",ninetieth_percentile)

    b = np.where(x<tenth_percentile, tenth_percentile, x)

    new_data = np.where(b>ninetieth_percentile, ninetieth_percentile, b)

    return new_data
```

```python
for column in df3.columns:
    print(column)
    try:
        a=floor_cap(df3[column])
        df3[column] = a
    except:
        pass
```

```python
x = df4.drop("violentPerPop", axis = 1)
y = df4['violentPerPop']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=0)

X_train.shape, X_test.shape
```

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature
cor = X_train.corr()
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.85)
len(set(corr_features))
```

```python
corr_features
```

```python
X_train.drop(corr_features, axis = 1, inplace=True)
X_test.drop(corr_features, axis = 1, inplace=True)
```

# Feature Scaling

### We are using Normalization or Standardization method to scale our dataset.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Standarization rescales value in such a way that its mean will be 0 and standard deviation will be 1
```

```python
scale = StandardScaler()
X_train_scaled=scale.fit_transform(X_train)
X_test_scaled=scale.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
```

```python
#Linear Regression
reg = LinearRegression().fit(X_train_scaled, y_train)

predict = reg.predict(X_test_scaled)
lr_mae = mae(y_test,predict)
lr_mse = mse(y_test,predict)
lr_rmse = mse(y_test,predict,squared = False)
lr_mae, lr_mse, lr_rmse
```

```python
reg = DecisionTreeRegressor(max_depth = 20).fit(X_train_scaled, y_train)

predict = reg.predict(X_test_scaled)
dt_mae = mae(y_test,predict)
dt_mse = mse(y_test,predict)
dt_rmse = mse(y_test,predict,squared = False)
dt_mae, dt_mse, dt_rmse
```

```python
reg = RandomForestRegressor(max_depth=35).fit(X_train_scaled,y_train)

predict = reg.predict(X_test_scaled)
rf_mae = mae(y_test,predict)
rf_mse = mse(y_test,predict)
rf_rmse = mse(y_test,predict,squared = False)
rf_mae, rf_mse, rf_rmse
```

```python
model_results = pd.DataFrame({'Model':['LinearRegression','DecicisonTree','RandomForest'],
                            'MAE' : [lr_mae,dt_mae,rf_mae],
                            'MSE' : [lr_mse,dt_mse,rf_mse],
                             'RMSE' : [lr_rmse,dt_rmse,rf_rmse]})
model_results
```
